```python
import requests
import folium
from polyline import decode  # pip install polyline
import os

# Google Routes APIのエンドポイント
url = "https://routes.googleapis.com/directions/v2:computeRoutes"

# APIキー
api_key = os.getenv("Google_API_Key")

# リクエストヘッダー
headers = {
    "Content-Type": "application/json",
    "X-Goog-Api-Key": api_key,
    "X-Goog-FieldMask": "routes.duration,routes.distanceMeters,routes.polyline.encodedPolyline"
}

# 出発地・到着地・交通手段
payload = {
    "origin": {
        "location": {
            "latLng": {
                "latitude": 35.665251,
                "longitude": 139.712092
            }
        }
    },
    "destination": {
        "location": {
            "latLng": {
                "latitude": 35.661971,
                "longitude": 139.703795
            }
        }
    },
    "travelMode": "WALK"
}

# POSTリクエストでRoutes APIを呼び出し
response = requests.post(url, headers=headers, json=payload)
data = response.json()

# ルートデータを取得してpolylineをデコード
encoded_polyline = data["routes"][0]["polyline"]["encodedPolyline"]
route = decode(encoded_polyline)  # [(lat, lng), ...]

# foliumで描画
m = folium.Map(location=route[0], zoom_start=15)
folium.Marker(route[0], tooltip="出発", icon=folium.Icon(color="green")).add_to(m)
folium.Marker(route[-1], tooltip="到着", icon=folium.Icon(color="red")).add_to(m)
folium.PolyLine(route, color="blue", weight=5).add_to(m)

m.save("google_routes_map.html")

m
```
